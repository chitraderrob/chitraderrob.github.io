```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# File to Load 
cities_data = "cities.csv"

# Convert cities_data to data frame
cities_df = pd.read_csv(cities_data)
cities_df.head()
```

```python
# Format Data Frame
cities_df_new=cities_df.set_index('City_ID')
cities_df_new=cities_df_new[['City', 'Country', 'Date', 'Lat', 'Lng', 'Max Temp', 'Humidity', 'Cloudiness', 'Wind Speed']]
cities_df_new['Date']=pd.to_datetime(cities_df_new['Date'], unit='s')
cities_df_new['Date'] = cities_df_new['Date'].dt.date
cities_df_new.head()
```

```python
# Set date as variable
weather_date=cities_df_new.iloc[0,2].strftime("%m/%d/%Y")
weather_date
```

# Latitude vs. Temp

```python
plt.scatter(cities_df_new['Lat'], cities_df_new['Max Temp'],marker="o", color="blue", edgecolors="black", alpha=0.75)
plt.title(f"City Latitude vs. Max Temperature ({weather_date})")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.grid()
plt.savefig("static/images/Lat_vs_Temp.png")
plt.show()
```

# Latitude vs Humidity

```python
plt.scatter(cities_df_new['Lat'], cities_df_new['Humidity'],marker="o", color="blue", edgecolors="black", alpha=0.75)
plt.title(f"City Latitude vs. Humidity ({weather_date})")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.grid()
plt.savefig("static/images/Lat_vs_Humidity.png")
plt.show()
```

# Latitude vs Cloudiness

```python
plt.scatter(cities_df_new['Lat'], cities_df_new['Cloudiness'],marker="o", color="blue", edgecolors="black", alpha=0.75)
plt.title(f"City Latitude vs. Cloudiness ({weather_date})")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.grid()
plt.savefig("static/images/Lat_vs_Cloudiness.png")
plt.show()
```

# Latitude vs Wind Speed

```python
plt.scatter(cities_df_new['Lat'], cities_df_new['Wind Speed'],marker="o", color="blue", edgecolors="black", alpha=0.75)
plt.title(f"City Latitude vs. Wind Speed ({weather_date})")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.grid()
plt.savefig("static/images/Lat_vs_Wind_Speed.png")
plt.show()
```
